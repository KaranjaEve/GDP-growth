# Project: GDP Growth Analysis- European Union vs Euro Area 2008 -2024

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option("display.max_columns",500)
plt.style.use("fivethirtyeight")
```

# 1. Import EU data

```python
import pandas as pd

gdp_growth = pd.read_csv("data/GDP growth.csv",skiprows=3)
#view the DataFrame
gdp_growth.head()
```

# 2. Clean raw data

```python

#check columns
gdp_growth.columns
```

```python
#drop unnecessary columns
gdp_growth = gdp_growth.drop(columns=['Unnamed: 69'])
gdp_growth
```

```python
#change the view for better visualisation
gdp_growth = gdp_growth.melt(
    id_vars=['Country Name', 'Country Code','Indicator Name','Indicator Code'],
    var_name= 'Year',
    value_name='Value'
             
)
gdp_growth.head()
```

```python
#drop missing values
gdp_growth= gdp_growth.dropna(subset=['Value'])
gdp_growth.head()
```

```python
#ensure Year column is int
gdp_growth.dtypes
```

```python
#change type of Year
gdp_growth["Year"] = gdp_growth["Year"].astype(int)

gdp_growth.dtypes
```

# 3. Minimize search to EU and Euro Area

```python
gdp_growth_eu_ea = gdp_growth.query("`Country Name` in ['European Union', 'Euro area' ] and Year >=2008" )
gdp_growth_eu_ea.head()
```

```python
#check for missing values
gdp_growth_eu_ea.isnull().sum()
```

```python
#our data decription
gdp_growth_eu_ea.describe()
```

# 4. Create visualisations

```python


plt.figure(figsize=(12,5))
countries = gdp_growth_eu_ea["Country Name"].unique()
years = sorted(gdp_growth_eu_ea["Year"].unique())

# Create an array for x positions
x = np.arange (len(years))
bar_width = 0.35

for i, country in enumerate(countries):
    group = gdp_growth_eu_ea[gdp_growth_eu_ea["Country Name"] == country].sort_values("Year")
    plt.bar(x + i * bar_width, group["Value"], width=bar_width, alpha=0.7, label=country, edgecolor='black')

plt.axhline(0, color='black', linewidth=0.5)

for spine in plt.gca().spines.values():
    spine.set_visible(True)
    spine.set_color('black')
    spine.set_linewidth(1.5)

plt.xticks(x + bar_width / 2, years, fontsize=9)  # labels the x-axis with years centered at the bars
plt.xlabel("Year")
plt.ylabel("GDP Growth rate (%)")
plt.title("GDP Growth rate: EU vs Euro Area")
plt.legend()
plt.grid(False)
plt.grid(axis = 'y')
plt.tight_layout()
plt.savefig("figures/GDP Growth rate.png")
plt.show()
```
